```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file, handling commas as thousand separators
df = pd.read_csv('Biological_Process.csv', thousands=',')

# Set 'Description' as the index to structure the data correctly
df.set_index('Description', inplace=True)

# Calculate relative abundance (percentage of total for each category)
df_relative = (df.div(df.sum(axis=0)) * 100).round(2)

# Transpose the DataFrame to prepare for plotting
df_plot = df_relative.T

# Create a stacked bar chart
plt.figure(figsize=(12, 10))
ax = df_plot.plot(kind='bar', stacked=True, colormap='tab20', edgecolor='black')

# Customize labels and title
plt.title('Relative Abundance of Biological Processes by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Relative Abundance (%)', fontsize=12)

# Adjust legend to show all processes outside the plot
plt.legend(
    title='Biological Process',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    fontsize='small'
)

# Format y-axis to show percentage
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

# Ensure layout is tight to accommodate the legend
#plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read and preprocess data
df = pd.read_csv('Biological_Process.csv', thousands=',')
df.set_index('Description', inplace=True)

# Calculate relative percentages
df_relative = (df.div(df.sum(axis=0)) * 100).round(2)

# Create grouped bar plot
plt.figure(figsize=(6, 14))
ax = df_relative.plot(kind='bar', 
                     colormap='tab20',
                     width=0.85,
                     edgecolor='black')

# Customize appearance
plt.title('Gene Ontology - Biological Processes', pad=20, fontsize=14)
plt.xlabel('Biological Process', fontsize=12)
plt.ylabel('Relative Abundance (%)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', alpha=0.3)

# Formatting tweaks
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))
plt.legend(title='Category', 
          bbox_to_anchor=(1.02, 1),
          loc='upper left',
          frameon=False)
# Save the figure in TIFF format with high resolution
output_filename = "Biological_Processes.tiff"
plt.savefig(output_filename, format='tiff', dpi=300, bbox_inches='tight')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file names and titles
files = ['Biological_Process.csv', 'Molecular_Functions.csv', 'Cellular_Components.csv']
titles = ['(a)', '(b)', 'c']
output_files = ['Biological_Processes.tiff', 'Molecular_Functions.tiff', 'Cellular_Components.tiff']

# Create a figure with three subplots
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(6, 18), constrained_layout=True)

# Loop through the files and create a subplot for each
for i, (file, title, output_file) in enumerate(zip(files, titles, output_files)):
    # Read and preprocess data
    df = pd.read_csv(file, thousands=',')
    df.set_index('Description', inplace=True)
    
    # Calculate relative percentages
    df_relative = (df.div(df.sum(axis=0)) * 100).round(2)
    
    # Create grouped bar plot
    ax = df_relative.plot(kind='bar', colormap='tab20', width=0.85, edgecolor='black', ax=axes[i])
    
    # Customize appearance
    ax.set_title(title, pad=20, fontsize=14)
    ax.set_xlabel('')  # Remove x-label to avoid clutter
    ax.set_ylabel('Relative Abundance (%)', fontsize=12)
    ax.tick_params(axis='x', rotation=80, labelsize=8)
    ax.grid(axis='y', alpha=0.3)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))
    ax.legend(title='Category', bbox_to_anchor=(1.02, 1), loc='upper left', frameon=False)
    
    # Save each subplot as a separate TIFF file
    fig.savefig(output_file, format='tiff', dpi=300, bbox_inches='tight')

# Adjust layout and show the plot
plt.show()
```
